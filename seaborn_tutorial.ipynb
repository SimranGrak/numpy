```python
#seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
print(type(tips))
print(tips.head())
```

```python
sns.set_theme(style=("whitegrid"))

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size",style="smoker"
)
```

```python
sns.set_theme(style=("whitegrid"))

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    kind="line",
    size="size",style="smoker"
)
```

```python
flights=sns.load_dataset("flights")

sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)

```

```python
print(flights.head())
```

```python
print(tips.head())
```

```python
sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
sns.boxplot(
    data=tips,
    x="day",
    y="tip",
)
```

```python
penguins=sns.load_dataset("penguins")

sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
)
```

```python
print(penguins.head())
```

```python
flights_pivot=flights.pivot(index="month",columns="year", values="passengers")

sns.heatmap(
    flights_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d"
)

plt.title("passengers heatmap")
plt.tight_layout()
```
