```python
import json
```

```python
#load the data
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
        
    return data
```

```python
data=load_data("store_data.json")
print(data)
print(type(data))
```

```python
#clean and structure data
def clean_data(data):
    text_to_num={"one":1, "two":2, "three":3, "four":4, "five":5}
    cleared_data=[]
    unique_users=set()
    
    for user in data:
        #clean rating
        raw_rating=str(user["rating"]).strip().lower()
        if (raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]
            
        #handling missing vals
        raw_age=user.get("age")
        if (raw_age== None):
            user["age"]=None

        #dedupliaction
        if (user["name"] in unique_users):
            continue

        unique_users.add(user["name"])
        cleared_data.append(user)

    return cleared_data


```

```python
data=clean_data(data)
```

```python
#get meaningful insights
def get_insight(data):
    #avg_rating
    tot_rating=0
    for user in data:
        tot_rating+=float(user["rating"])

    print(f"average rating: {tot_rating/len(data)}")

    #percentage of poor rating
    poor_rating=0
    for user in data:
        if (float(user["rating"])<3):
            poor_rating+=1

    print(f"percentage of poor rating: {poor_rating/len(data)*100}")
```

```python
get_insight(data)
```

```python
#build recommendation feature
def get_recommendation(data):
    recommendation=[]

    for user in data:
        curr_recomm={}
        curr_recomm["name"]= user["name"]
        if (float(user["rating"])>=4):
            curr_recomm["brand"]="Apple"
        else:
            curr_recomm["brand"]="Samsung"
        recommendation.append(curr_recomm)

    return recommendation

    
```

```python
get_recommendation(data)
```
