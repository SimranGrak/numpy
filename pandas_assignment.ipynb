```python
#Pandas assignement
import pandas as pd
```

```python
#Question 1: Start by loading the dataset.
#A Display the following information:
#i First 10 rows
#ii Shape & Data types
#iii Summary statistics(mean,std,min,max)

df=pd.read_csv("IRIS.csv")

#i part
ten_rows=df.head(10)
ten_rows


#ii part
print(df.dtypes)
print(df.shape)


#iii part
summary=df.agg({
    "sepal_length":[ "mean", "std", "min", "max"],
    "sepal_width":["mean", "std", "min", "max"],
    "petal_length":["mean", "std", "min", "max"],
    "petal_width": ["mean", "std", "min", "max"]
})

print(summary)


#B Select only those rows where:
#i petal_length>4.5
#ii species="Iris-virginica"


#i part
petal_leng=df.copy()
petal_leng= df[df["petal_length"]> 4.5]
print(petal_leng)


#ii part
species=df.copy()
species=df[df["species"]=="Iris-virginica"]
print(species)


#C Group by species and compute:
#i average sepal length 
#ii Maximum petal_width
#iii standard deviation of sepal_width


#i part
avg_seplen=df.copy()
avg_seplen=df.groupby("species")["sepal_length"].mean()
print(avg_seplen)


#ii part
max_petwid=df.copy()
max_petwid=df.groupby("species")["petal_width"].max()
print(max_petwid)


#iii part
std_petwid=df.copy()
std_petwid=df.groupby("species")["sepal_width"].std()
print(std_petwid)


#D Create a new column: "petal_ratio"=petal_length/petal_width
#For each species, find the average petal_ratio

df2=df.copy()
df2=df2.assign(petal_ratio=df2["petal_length"]/df2["petal_width"])
df2
```

```python
#Question 2: Start by loading the TITANIC dataset.
#A For each species: Name, sex, Age, Fare, Survived

#B Select passengers who:
#i Are female
#ii Fare> 30

#C Group by Pclass and compute:
#i Survival rate
#ii Average Fare
#iii Average Age


#A part
df=pd.read_csv("Titanic-Dataset.csv")
selected_column=df.copy()
selected_column=df[["Name", "Sex", "Age", "Fare", "Survived"]]
print(selected_column)


#B part
#i part
only_female=df.copy()
only_female=df[df["Sex"]=="female"]
print(only_female)


#ii part
select_fare=df.copy()
select_fare=df[df["Fare"]>30]
print(select_fare)


#C part
#i part
survival_rate=df.groupby("Pclass")["Survived"].mean()
print(survival_rate)

#ii part
avg_fare=df.groupby("Pclass")["Fare"].mean()
print(avg_fare)

#iii part
avg_age=df.groupby("Pclass")["Age"].mean()
print(avg_age)
```
