```python
import json
import os

# JSON Module
'''User Data Storage
What to do
1. Store multiple users with:
    a) id
    b) name
    c) email
2. Save users into a JSON file
3. Read users back and display them'''

print("current directory:", os.getcwd())

data={
    "id":1,
    "name":"simran",
    "email":"simrangrak1761@gmail.com"
}

print("before writing")
with open("data.json", "w")as f:
    json.dump(data, f, indent=4)

print("after writing")


with open("data.json", "r")as f:
    df=json.load(f)
    print(df)
```
