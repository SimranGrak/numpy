```python
#Data Filtering
'''Data Filtering
What to do
a) Read numbers from a file
b) Create:
    1. List of even numbers
    2. List of numbers > 50
    3.Square of all numbers'''

with open("notes.txt", "r")as f:
    numbers=[int(line.strip()) for line in f if line.strip()]
    print(numbers)


even=[x for x in numbers if x%2==0]
print(even)

greater=[x for x in numbers if x>50]
print(greater)

square=[x*x for x in numbers]
print(square)
```
