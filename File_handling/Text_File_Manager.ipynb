```python
#Text File Manager
'''What to do
a) Create a text file named notes.txt
Allow user to:
b) Add a new note
c) View all notes
d) Search a note by keyword'''

user=input("enter notes:")
with open("notes.txt", "a")as f:
    data=f.write(user + "\n")
    print(data)

keyword=input("Enter keyword:")
with open("notes.txt", "r")as f:
    content=f.read()
    if keyword in content:
        print("yes, matched")

    else:
        print("not, matched")
```
