```python
#Student Record System
'''
1. What to build
    a) Add student
    b) View all students
    c) Search by roll number
    d) Delete student
2. Rules
    a) Store data in JSON file
    b) Use exception handling everywhere
    c) Use list comprehension for filtering/searching
    d) Use File I/O only (no database)'''



#Add student
import json
import os

data=[
    {
        "roll_no":1,
        "name":"simran",
        "city":"karnal"
    },
    {
        "roll_no":2,
        "name":"suman",
        "city":"noida"
    },
      {
        "roll_no":3,
        "name":"punjab",
        "city":"canada"
    }
]

try:
    with open("students.json", "w")as f:
        df=json.dump(data, f, indent=4)
        print("student added successfully!")
except PermissionError:
    print("do not have permission to write in this file!")
    
```

```python
#View all students

class EmptyFile(Exception):
    pass


try:
    with open("students.json", "r")as f:
        df=f.read()
        if not df.strip():
            raise EmptyFile("empty file")

        students=json.loads(df)

    print("all students:")
    for student in students:
        print(student)
    

except EmptyFile as e:
    print("error:", e)
```

```python
#Search by roll number
try:
    keyword=int(input("enter the roll no: "))
    with open("students.json", "r")as f:
        students=json.load(f)

    found=False
    for student in students:
        if student["roll_no"]==keyword:
            print(student)
            found=True

    if not found:
        print("student not found")
        
except FileNotFoundError:
    print("file not found")
except ValueError:
    print("please enter a valid roll no")
```

```python
#Delete student
try:
    value=input("Enter the name of student:")
    with open("students.json", "r")as f:
        df=json.load(f)

    new_list=[]
    found=False
    for student in df:
        if student["name"]!=value:
            new_list.append(student)
        else:
            found=True

    if found:
        with open("students.json","w")as f:
            df=json.dump(new_list, f, indent=4)
        print("student successfully deleted!")
    else:
        print("student not found!")
except FileNotFound:
    print("file not found")
except json.JSONDecodeError:
    print("Invalid JSON data")
            
    
```
