```python
# Safe File Reader
'''What to do
a) Ask user for a filename
b) Try to read it
c) Handle:
    1.File not found
    2.Empty file
    3.Permission error'''

#File not found
try:
    with open("notes.txt", "r")as f:
        content=f.read()
        print(content)
except FileNotFoundError:
    print("File not found")
```

```python
#Empty file

class EmptyFile(Exception):
    pass

try:
    with open("content.txt", "r")as f:
        content=f.read()
        if not content.strip():
            raise EmptyFile("No content!")
except EmptyFile as e:
    print("Error:", e)
finally:
    print("compeleted")
```

```python
#Permission Error
try:
    with open("C:/Windows/system32/test.txt", "w")as f:
        f.write("hello")
except PermissionError:
    print("you do not have permission to write in this file")
finally:
    print("complete")
```
